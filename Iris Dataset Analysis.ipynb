```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/kaggle/input/iris-dataset/IRIS.csv')
data.head(10)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['species'].unique()
```

```python
data.describe()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,6))
sns.barplot(x='species', y='sepal_length', data=data, ci='sd', palette='pastel')
plt.title("Average Sepal Length by Species", fontsize=14)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Average Sepal Length (cm)", fontsize=12)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='sepal_width', data=data, ci='sd', palette="muted")
plt.title("Average Sepal Width by Species", fontsize=14)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Average Sepal Width (cm)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal_length', data=data, ci='sd', palette="dark")
plt.title("Average Petal Length by Species", fontsize=14)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Average Petal Length (cm)", fontsize=12)
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal_width', data=data, ci='sd', palette="bright")
plt.title("Average Petal Width by Species", fontsize=14)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Average Petal Width (cm)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['sepal_length'], bins=15, color='red', alpha=0.7, edgecolor='black')
plt.title("Distribution of Sepal Length", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['sepal_width'], bins=15, color='blue', alpha=0.7, edgecolor='black')
plt.title("Distribution of Sepal Length", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['petal_length'], bins=15, color='orange', alpha=0.7, edgecolor='black')
plt.title("Distribution of Sepal Length", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['sepal_width'], bins=15, color='violet', alpha=0.7, edgecolor='black')
plt.title("Distribution of Sepal Length", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=data, hue='species', palette='muted')
plt.title("Sepal Length vs Sepal Width", fontsize=14)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Sepal Width (cm)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal_length', y='petal_width', data=data, hue='species', palette='muted')
plt.title("Petal Length vs Petal Width", fontsize=14)
plt.xlabel("Petal Length (cm)", fontsize=12)
plt.ylabel("Petal Width (cm)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal_length', y='petal_width', data=data, hue='species', palette='muted')
sns.regplot(x='petal_length', y='petal_width', data=data, scatter_kws={'s': 20}, line_kws={'color': 'red'}, scatter=True, ci=90)
plt.title("Petal Length vs Petal Width", fontsize=14)
plt.xlabel("Petal Length (cm)", fontsize=12)
plt.ylabel("Petal Width (cm)", fontsize=12)
plt.show()
```

```python
correlation_matrix = data.drop('species', axis=1).corr()
sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.1)
plt.title("Correlation Heatmap of Features", fontsize=14)
plt.show()
```
